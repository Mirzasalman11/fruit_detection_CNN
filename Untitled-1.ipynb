....................................................................Data Argumentation .............................................................

```python
def data_arg (odir,ndir,ipc,tec):
    import tensorflow as tf
    from tensorflow.keras.preprocessing.image import ImageDataGenerator
    import os

    # Define the data augmentation parameters
    if(tec=='train'):
        datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        horizontal_flip=True,
        rotation_range=40,
        fill_mode='nearest')
    else:
        datagen = ImageDataGenerator(
        rescale=1./255,        
    )


    # Specify the directory where your original images are located
    original_data_dir = odir

    # Create a directory to store the augmented images
    augmented_data_dir = ndir
    if not os.path.exists(augmented_data_dir):
        os.makedirs(augmented_data_dir)

    # Set the desired number of augmented images per class
    images_per_class = ipc

    # Iterate through class subdirectories and generate augmented images
    for class_dir in os.listdir(original_data_dir):
        class_path = os.path.join(original_data_dir, class_dir)
        
        if not os.path.isdir(class_path):
            continue  # Skip non-directory files in the original_data_dir
        
        augmented_class_dir = os.path.join(augmented_data_dir, class_dir)
        if not os.path.exists(augmented_class_dir):
            os.makedirs(augmented_class_dir)

        # Initialize a counter for the number of augmented images generated
        generated_images = 0

        for root, _, files in os.walk(class_path):
            for file in files:
                img = tf.keras.preprocessing.image.load_img(os.path.join(root, file))
                x = tf.keras.preprocessing.image.img_to_array(img)
                x = x.reshape((1,) + x.shape)

                i = 0
                for batch in datagen.flow(x, batch_size=1):
                    augmented_image = tf.keras.preprocessing.image.array_to_img(batch[0])
                    augmented_image.save(os.path.join(augmented_class_dir, f'augmented_{i}_{file}'))
                    i += 1
                    generated_images += 1
                    if i >= images_per_class:
                        break

                if generated_images >= images_per_class:
                    break

            if generated_images >= images_per_class:
                break

```

```python
data_arg('archive/train','train',600,'train')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
model = keras.models.Sequential([    
    keras.layers.Conv2D(filters=16, kernel_size=(3,3), activation='relu', input_shape=(30,30,3)),
    keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'),
    keras.layers.MaxPool2D(pool_size=(2, 2)),
    keras.layers.BatchNormalization(axis=-1),
    
    keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    keras.layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'),
    # keras.layers.Conv2D(filters=512, kernel_size=(3,3), activation='relu'),

    keras.layers.MaxPool2D(pool_size=(2, 2)),
    keras.layers.BatchNormalization(axis=-1),

    keras.layers.Flatten(),


    keras.layers.Dense(512, activation='relu'),
    # keras.layers.Dense(128, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(rate=0.5),
    
    keras.layers.Dense(33, activation='softmax')
])
```

```python
class_names=['Apple Braeburn',
 'Apple Granny Smith',
 'Apricot',
 'Avocado',
 'Banana',
 'Blueberry',
 'Cactus fruit',
 'Cantaloupe',
 'Cherry',
 'Clementine',
 'Corn',
 'Cucumber Ripe',
 'Grape Blue',
 'Kiwi',
 'Lemon',
 'Limes',
 'Mango',
 'Onion White',
 'Orange',
 'Papaya',
 'Passion Fruit',
 'Peach',
 'Pear',
 'Pepper Green',
 'Pepper Red',
 'Pineapple',
 'Plum',
 'Pomegranate',
 'Potato Red',
 'Raspberry',
 'Strawberry',
 'Tomato',
 'Watermelon']
```

```python
import os
import cv2
import numpy as np
from PIL import Image

# Collecting the Training Data
image_data = []
image_labels = []


for i in class_names:
    path = 'train/' + i
    images = os.listdir(path)

    for img in images:

        image = cv2.imread(path + '/' + img)
        # Resize the image using cv2
        resize_image = cv2.resize(image, (30, 30))
        rgb_image = cv2.cvtColor(resize_image, cv2.COLOR_BGR2RGB)
        
        image_data.append(np.array(rgb_image))
        image_labels.append(i)


# Changing the list to numpy array
image_data = np.array(image_data)
image_labels = np.array(image_labels)

print(image_data.shape, image_labels.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(image_data, image_labels, test_size=0.3, random_state=42, shuffle=True)


print("X_train.shape", X_train.shape)
print("X_valid.shape", X_val.shape)
print("y_train.shape", y_train.shape)
print("y_valid.shape", y_val.shape)
```

```python
import matplotlib.pyplot as plt

# Iterate through the dataset and display a few images

plt.figure(figsize=(11, 9))
for i in range(25):  # Display the first 9 images from the batch
    ax = plt.subplot(5, 5, i + 1)
    plt.imshow(X_train[i])
    plt.title(y_train[i])
    plt.axis("off")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical

train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,  # Normalize pixel values
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

#to_categorical is used for one hote encoding of y

train_generator = train_datagen.flow(X_train, to_categorical(y_train_encoded))

validation_datagen = ImageDataGenerator(rescale=1.0/255.0)

validation_generator = validation_datagen.flow(X_val, to_categorical(y_val_encoded))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_generator, epochs=3, validation_data=validation_generator)
```
